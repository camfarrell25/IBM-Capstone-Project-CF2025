This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## Load the dataframe


Load the data


```python

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
ssl_context = ssl._create_unverified_context()

response = urllib.request.urlopen(url, context=ssl_context)
data = pd.read_csv(response)

data.head()

```

```python
url2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_3.csv"
ssl_context = ssl._create_unverified_context()

response2 = urllib.request.urlopen(url2, context=ssl_context)
X = pd.read_csv(response2)

X.head(100)
```

## TASK  1


Create a NumPy array from the column <code>Class</code> in <code>data</code>, by applying the method <code>to_numpy()</code>  then
assign it  to the variable <code>Y</code>,make sure the output is a  Pandas series (only one bracket df\['name of  column']).


```python
Y = pd.Series(data['Class'].to_numpy())
```

```python
#Check type of Y
print(type(Y))
#should return pandas.core.series.Series
```

## TASK  2


Standardize the data in <code>X</code> then reassign it to the variable  <code>X</code> using the transform provided below.


```python
# students get this 
transform = preprocessing.StandardScaler()
X = transform.fit_transform(X)
```

We split the data into training and testing data using the  function  <code>train_test_split</code>.   The training data is divided into validation data, a second set used for training  data; then the models are trained and hyperparameters are selected using the function <code>GridSearchCV</code>.


## TASK  3


Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to  0.2 and random_state to 2. The training data and test data should be assigned to the following labels.


<code>X_train, X_test, Y_train, Y_test</code>


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

we can see we only have 18 test samples.


```python
Y_test.shape
```

## TASK  4


Create a logistic regression object  then create a  GridSearchCV object  <code>logreg_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
# Provided parameter grid
parameters = {"C": [0.01, 0.1, 1], "penalty": ["l2"], "solver": ["lbfgs"]}

# Create logistic regression model
lr = LogisticRegression()

# Create GridSearchCV object
logreg_cv = GridSearchCV(lr, parameters, cv=10)

# Fit the model on training data
logreg_cv.fit(X_train, Y_train)
```

We output the <code>GridSearchCV</code> object for logistic regression. We display the best parameters using the data attribute <code>best_params\_</code> and the accuracy on the validation data using the data attribute <code>best_score\_</code>.


```python
print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
print("accuracy :",logreg_cv.best_score_)
```

## TASK  5


Calculate the accuracy on the test data using the method <code>score</code>:


```python
test_accuracy = logreg_cv.score(X_test, Y_test)
print("Test set accuracy:", test_accuracy)
```

Lets look at the confusion matrix:


```python
yhat=logreg_cv.predict(X_test)
plot_confusion_matrix(Y_test,yhat)
```

Examining the confusion matrix, we see that logistic regression can distinguish between the different classes.  We see that the problem is false positives.

False Postive - 3 (True label is not landed, Predicted label is landed)

## TASK  8


Create a decision tree classifier object then  create a  <code>GridSearchCV</code> object  <code>tree_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
parameters = {'criterion': ['gini', 'entropy'],
     'splitter': ['best', 'random'],
     'max_depth': [2*n for n in range(1,10)],
     'max_features': ['auto', 'sqrt'],
     'min_samples_leaf': [1, 2, 4],
     'min_samples_split': [2, 5, 10]}

tree = DecisionTreeClassifier()
```

## TASK  12


Find the method performs best:


```python
print("Logistic Regression Test Accuracy:", logreg_cv.score(X_test, Y_test))
print("SVM Test Accuracy:", sum_cv.score(X_test, Y_test))
print("Decision Tree Test Accuracy:", tree_cv.score(X_test, Y_test))
print("KNN Test Accuracy:", knn_cv.score(X_test, Y_test))
```

```python
from sklearn.metrics import classification_report

models = {
    "Logistic Regression": logreg_cv,
    "SVM": sum_cv,
    "Decision Tree": tree_cv,
    "K-Nearest Neighbors": knn_cv
}

for name, model in models.items():
    print(f"Classification Report for {name}:")
    y_pred = model.predict(X_test)
    print(classification_report(Y_test, y_pred))
    print("="*60)
```

```python
#Checking if predictions match across different models
preds = {name: model.predict(X_test) for name, model in models.items()}

for name1, pred1 in preds.items():
    for name2, pred2 in preds.items():
        if name1 != name2:
            print(f"Do predictions of {name1} and {name2} match? { (pred1 == pred2).all() }")
```

```python
models = {
    "Logistic Regression": logreg_cv,
    "SVM": sum_cv,
    "Decision Tree": tree_cv,
    "K-Nearest Neighbors": knn_cv
}

print("Model CV Scores Comparison:")
for name, model in models.items():
    print(f"{name}: Best CV Accuracy = {model.best_score_:.4f}")

```

```python
sorted_models = sorted(models.items(), key=lambda item: item[1].best_score_, reverse=True)

print("Models sorted by best CV accuracy:")
for name, model in sorted_models:
    print(f"{name}: {model.best_score_:.4f}")
```

```python
import matplotlib.pyplot as plt

# Dictionary of models and their best cross-validation score
model_scores = {
    'Logistic Regression': logreg_cv.best_score_,
    'SVM': sum_cv.best_score_,
    'Decision Tree': tree_cv.best_score_,
    'KNN': knn_cv.best_score_
}

# Identify best model
best_model = max(model_scores, key=model_scores.get)
best_score = model_scores[best_model]

# Bar colors
colors = ['skyblue' if model != best_model else 'mediumseagreen' for model in model_scores]

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(model_scores.keys(), model_scores.values(), color=colors)
plt.title('🔍 Cross-Validated Accuracy by Classification Model', fontsize=14)
plt.ylabel('Best CV Accuracy')
plt.ylim(0, 1.0)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height + 0.01, f"{height:.2f}", ha='center', va='bottom')

# Annotate best model
plt.text(
    list(model_scores.keys()).index(best_model),
    best_score + 0.05,
    f'🏆 Best: {best_model}\nAccuracy: {best_score:.2f}',
    ha='center',
    va='bottom',
    fontsize=11,
    fontweight='bold',
    color='darkgreen'
)

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
model_scores = {
    'Logistic Regression': logreg_cv.best_score_,
    'SVM': sum_cv.best_score_,
    'Decision Tree': tree_cv.best_score_,
    'KNN': knn_cv.best_score_
}
best_model_name = max(model_scores, key=model_scores.get)

# Predict on test data based on best model
if best_model_name == 'Logistic Regression':
    y_pred = logreg_cv.predict(X_test)
    model_used = logreg_cv
elif best_model_name == 'SVM':
    y_pred = sum_cv.predict(X_test)
    model_used = sum_cv
elif best_model_name == 'Decision Tree':
    y_pred = tree_cv.predict(X_test)
    model_used = tree_cv
elif best_model_name == 'KNN':
    y_pred = knn_cv.predict(X_test)
    model_used = knn_cv

# Plot confusion matrix
ConfusionMatrixDisplay.from_estimator(model_used, X_test, Y_test, cmap='Blues')
plt.title(f"Confusion Matrix for {best_model_name}")
plt.show()
```
